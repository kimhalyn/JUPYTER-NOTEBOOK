```python
# 실습테스트 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100) # 0~10의 범위 안에서 100개의 값 생성
```

```python
plt.figure() # 새로운 그래프 창 생성
plt.plot(x, np.sin(x), 'b-')
plt.plot(x, np.cos(x), 'y--')
plt.show()
```

```python
# 실습테스트 2
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.show()
```

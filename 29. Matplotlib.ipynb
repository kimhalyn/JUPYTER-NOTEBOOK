```python
# 막대그래프
member_IDs = ['m1', 'm2', 'm3', 'm4'] # 회원 ID
before_ex = [45, 35, 42, 38] # 운동 시작 전 팔굽혀펴기 횟수
after_ex = [49, 42, 49, 37] # 운동 한 달 후 팔굽혀펴기 횟수
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_data = len(member_IDs) # 회원이 4명이므로 전체 데이터 수는 4
index = np.arange(n_data) # numpy를 이용하여 (0, 1, 2, 3) 배열 생성
plt.bar(index, before_ex) # bar(x,y)에서 x=index, y=before_ex로 지정
plt.show()
```

```python
# 라벨 생성
plt.bar(index, before_ex, tick_label= member_IDs)
plt.show()
```

```python
# 막대 색상 변경
plt.bar(index, before_ex, color=colors, tick_label=member_IDs)
plt.show()
```

```python
colors = ['r', 'g', 'k', 'm']
plt.bar(index, before_ex, color=colors, tick_label=member_IDs)
plt.show()
```

```python
# 막대 너비 조절
plt.bar(index, before_ex, color=colors, tick_label=member_IDs, width=0.3) # 기본값 0.8
plt.show()
```

```python
# 가로 막대로 표현
plt.barh(index, before_ex, color=colors, tick_label=member_IDs)
plt.show()
```

```python
barwidth = 0.4
plt.bar(index, before_ex, color='c', align='edge', width = barwidth, label = 'before')
plt.bar(index + barwidth, after_ex, color = 'm', align = 'edge', width = barwidth, label = 'after')
plt.show()
```

```python
barwidth = 0.4
plt.bar(index, before_ex, color='c', align='edge', width = barwidth, label = 'before')
plt.bar(index + barwidth, after_ex, color = 'm', align = 'edge', width = barwidth, label = 'after')

plt.xticks(index + barwidth, member_IDs)
plt.legend()
plt.xlabel('회원 ID')
plt.ylabel('팔굽혀 펴기 횟수')
plt.title('운동 시작 전과 후의 근지구력 변화 비교')
plt.show()
```

```python
# 파이 그래프
fruit = ['사과', '바나나', '딸기', '오렌지', '포도']
result = [6, 7, 4, 2, 1]
```

```python
import matplotlib.pyplot as plt
plt.pie(result)
plt.show()
```

```python
# 파이 사이즈 조절
plt.figure(figsize=(5, 5))
plt.pie(result)
plt.show()
```

```python
# 문자열 및 비율 지정
# 파이 사이즈 조절
plt.figure(figsize=(5, 5))
plt.pie(result, labels = fruit, autopct = '%.1f%%') # 소수점 한 자리로 표현
plt.show()
```

```python
# 시작 각 설정 및 방향 수정
# 파이 사이즈 조절
plt.figure(figsize=(5, 5))
plt.pie(result, labels = fruit, autopct = '%.1f%%', startangle=90, counterclock=False) # 시계방향으로 설정
plt.show()
```

```python
# 특정 값 강조

explode_value = (0.3, 0, 0, 0, 0)
plt.figure(figsize=(5, 5))
plt.pie(result, labels = fruit, autopct = '%.1f%%', startangle=90, counterclock=False, explode = explode_value, shadow=True) 
plt.show()
```
